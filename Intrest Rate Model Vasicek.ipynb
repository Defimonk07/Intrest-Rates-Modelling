```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Vasicek model
mu = 0.05      # Long-term mean interest rate
theta = 0.1    # Speed of reversion
sigma = 0.02   # Volatility
R0 = 0.03      # Initial interest rate
T = 10         # Total time in years
dt = 1/252     # Time step (daily)
N = int(T / dt)  # Number of time steps

# Initialize the array to store interest rates
R = np.zeros(N)
R[0] = R0

# Simulate the Vasicek model
for t in range(1, N):
    dW = np.random.normal(0, np.sqrt(dt))  # Wiener process
    R[t] = R[t-1] + theta * (mu - R[t-1]) * dt + sigma * dW

# Plot the interest rates
plt.figure(figsize=(10, 6))
plt.plot(np.linspace(0, T, N), R)
plt.title("Vasicek Model - Simulated Interest Rates")
plt.xlabel("Time (Years)")
plt.ylabel("Interest Rate")
plt.grid()
plt.show()
```
